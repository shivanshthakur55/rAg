## Rag on PDF

#### Extract text from pdf

```python
from langchain_community.document_loaders import PyPDFLoader
pdf_path = "Docs/LangChain.pdf"
loader =PyPDFLoader(pdf_path)
docs=loader.load()
docs
```

```python
for i in docs:
    i.metadata={"source":"LangChain","developer":"Lang"}
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter =RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=200
)
chunks=splitter.split_documents(docs)
chunks
```

```python
len(chunks)
```

#### Vector Store


```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

from langchain_community.vectorstores import Chroma
vectorstore=Chroma(
    embedding_function=embedding_model,
    persist_directory="./Vector"
)
```

```python
vectorstore.add_documents(chunks)
```

```python
vectorstore.similarity_search("what is langchain?",k=1)
```

#### Reuse the vector DB

```python
vectorstore_persist=Chroma(
    persist_directory="./Vector/",
    embedding_function=embedding_model
)
```

```python
vectorstore_persist.similarity_search("what is this document about?")
```
